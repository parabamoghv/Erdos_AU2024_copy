# Plotting Tips

In this notebook we provide some tips on how to improve your presentation plots with python.

```python
from sklearn.datasets import load_iris

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from seaborn import set_style
set_style("whitegrid")
```

## Axis and tick labels

Good plots have large, clean and easy to read axis and tick labels (when appropriate).

We will demonstrate what we mean by this using our iris data set.

```python
iris = load_iris(as_frame=True)

iris_df = iris['data']
iris_df['iris_class'] = iris['target']
```

```python
iris_df.head()
```

Let's plot sepal length against sepal width.

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df['sepal width (cm)'], iris_df['sepal length (cm)'])


plt.show()
```

The above plot has no axis labels and small, difficult to read tick labels.

In general, your plot should label its horizontal and vertical axes so your audience knows what it is they are looking at. When you add labels you should make them large enough to be read easily during your presentation.

Let's add those now. 

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df['sepal width (cm)'], iris_df['sepal length (cm)'])

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)


plt.show()
```

Just like we should have easy to read axis labels it is also preferable to have easy to read tick labels. This can also be accomplished with python.

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df['sepal width (cm)'], iris_df['sepal length (cm)'])

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

## adjusts the size of the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)


plt.show()
```

<i>Note that this is how you add and alter labels when you create your figure using `plt.figure`, however, if you you used `plt.subplots` you will want:</i>
- `ax.set_xlabel`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html</a>, 
- `ax.set_ylabel`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html</a>,
- `ax.set_xticks`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xticks.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xticks.html</a>, 
- `ax.set_xticklabels`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html</a>,
- `ax.set_yticks`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_yticks.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_yticks.html</a> and, 
- `ax.set_yticklabels`, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_yticklabels.html">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.set_yticklabels.html</a>.

## Informative legends

When appropriate you should add informative legends. Here informative means informative for your audience. 

For example, we have three types of iris, here is one example of a legend classifying the points by iris type.

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            label = "$y=0$")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            label = "$y=1$")
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            label = "$y=2$")

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

## adjusts the size of the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

## adds a legend
plt.legend(fontsize=12)


plt.show()
```

While this is certainly true, the labels $y=1$, $y=2$ and $y=3$ are not meaningful in terms of the original problem, classifying <i>irises</i>.

Here is a better legend.

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            label = "Setosa")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            label = "Versicolor")
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            label = "Virginica")

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

## adjusts the size of the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

## adds a legend
plt.legend(fontsize=12)


plt.show()
```

We should note that the position of the legend within the plot can be controlled with a `loc` argument, <a href="https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.legend">https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.axes.Axes.legend</a>.

## Appropriately using colors and markers

When using color or different marker types, make sure they are conveying information <i>not</i> already present in the plot. In the example above, we used color to convey the type of iris represented by each dot, which was not already present in the plot.

Let's look at an example from `seaborn`.

```python
from seaborn import swarmplot
```

```python
plt.figure(figsize=(8,6))

swarmplot(data = iris_df,
             x='iris_class',
             y='sepal width (cm)',
             edgecolor='black',
             linewidth=.5)


plt.show()
```

In this `swarmplot`, <a href="https://seaborn.pydata.org/generated/seaborn.swarmplot.html">https://seaborn.pydata.org/generated/seaborn.swarmplot.html</a> the color of the points duplicates information provided by the horizontal axis, that is the `iris_class` of the observation. We can thus remove it and have effectively the same plot.

```python
plt.figure(figsize=(8,6))

swarmplot(data = iris_df,
             x='iris_class',
             y='sepal width (cm)',
             color='dodgerblue',
             edgecolor='black',
             linewidth=.5)


plt.show()
```

#### Adding markers and color

In instances when you do encode information with color (like our above sepal length against sepal width plot) you may also consider changing the style of the marker or line as well. For example, consider this slightly altered sepal length against sepal width plot.

```python
plt.figure(figsize=(10,8))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            label = "Setosa")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            label = "Versicolor",
            marker='v')
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            label = "Virginica",
            marker='x')

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

## adjusts the size of the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

## adds a legend
plt.legend(fontsize=12)


plt.show()
```

While this example may seem contradictory to what I said above about duplicating information that already existed in the plot, the reason you may want to do this is to increase the accessibility of your figures. Not everyone can see differences in color or hue between the three classes. By adding a unique marker type to each class of iris you increase the number of people that can understand your plot as you intended.

For the different named `matplotlib` color options see this link, <a href="https://matplotlib.org/stable/gallery/color/named_colors.html">https://matplotlib.org/stable/gallery/color/named_colors.html</a>.

For the different scatter plot marker types see this link, <a href="https://matplotlib.org/stable/api/markers_api.html">https://matplotlib.org/stable/api/markers_api.html</a>.

For the different `matplotlib` line styles see this link, <a href="https://matplotlib.org/3.5.0/gallery/lines_bars_and_markers/linestyles.html">https://matplotlib.org/3.5.0/gallery/lines_bars_and_markers/linestyles.html</a>.

## Do not always stick to defaults

Let's return to our `swarmplot` example.

```python
plt.figure(figsize=(8,6))

swarmplot(data = iris_df,
             x='iris_class',
             y='sepal width (cm)',
             edgecolor='black',
             linewidth=.5)


plt.show()
```

There are several issues with the default values for `seaborn`'s `swarmplot` that could be improved upon:

- The default horizontal axis label is `iris_class` which comes from the `DataFrame` column name. Our audience would prefer a more human readable option, perhaps something like `Iris Type`.
- Similarly the default horizontal tick markers use the values from the `DataFrame` and not the real names of the iris types. We should change that.
- The default fontsizes are pretty small and difficult to read and
- We already mentioned changing the color to not duplicate information.

```python
plt.figure(figsize=(8,6))

swarmplot(data = iris_df,
             x='iris_class',
             y='sepal width (cm)',
             color='dodgerblue',
             edgecolor='black',
             linewidth=.5)

## Change the default xlabel
plt.xlabel("Iris Type", fontsize=12)

## Increase the ylabel fontsize
plt.ylabel("Sepal Width (cm)", fontsize=12)

## change the ytick sizes
plt.yticks(fontsize=10)

## change the xticks
plt.xticks([0,1,2], ["Setosa", "Versicolor", "Virginica"], fontsize=10)


plt.show()
```

The key takeaway here is that you should not always go with the default values for any given plotting function. Sometimes your plot can be drastically improved by setting options yourself.

## Saving high resolution plots

In a presentation setting you want to make sure your plots are easily read by your audience. One important step toward that goal is making your plots large.

One way to do this is to take your image and scale it up in the presentation slide creation software. However, this is not ideal because stretching your image to make it bigger can blur and distort some of the features of your plot.

A preferable approach is to save a large, high resolution version of your image and then scale it down in the presentation slide creation software. 

Here we will show you how to do that in `matplotlib`.

```python
plt.figure(figsize=(8,6))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            label = "Setosa")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            label = "Versicolor",
            marker='v')
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            label = "Virginica",
            marker='x')

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

## adjusts the size of the tick labels
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

## adds a legend
plt.legend(fontsize=12)


plt.show()
```

Recall our favorite sepal length against sepal width plot. 

The first thing we can do to save a large image is increase the `figsize` argument.

The second thing we can do is save the image in a higher resolution format like a `pdf` file or an `eps` image.

```python
## increase the figsize
plt.figure(figsize=(20,16))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            s=100,
            label = "Setosa")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            s=100,
            label = "Versicolor",
            marker='v')
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            s=100,
            label = "Virginica",
            marker='x')

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=22)
plt.ylabel("Sepal Length (cm)", fontsize=22)

## adjusts the size of the tick labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

## adds a legend
plt.legend(fontsize=16)

## saving the photo as a pdf
plt.savefig("iris_sepal_length_width.pdf",
               format='pdf')

### Alternatively ###

## saving the photo as an eps
plt.savefig("iris_sepal_length_width.eps",
               format='eps')


plt.show()
```

When you look at the save image files you may notice a ton of whitespace around the image. We can eliminate that with `matplotlib`'s `tight_layout` function, <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html">https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html</a>.

```python
## increase the figsize
plt.figure(figsize=(20,16))


plt.scatter(iris_df.loc[iris_df.iris_class==0, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==0, 'sepal length (cm)'],
            s=100,
            label = "Setosa")
plt.scatter(iris_df.loc[iris_df.iris_class==1, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==1, 'sepal length (cm)'],
            s=100,
            label = "Versicolor",
            marker='v')
plt.scatter(iris_df.loc[iris_df.iris_class==2, 'sepal width (cm)'], 
            iris_df.loc[iris_df.iris_class==2, 'sepal length (cm)'],
            s=100,
            label = "Virginica",
            marker='x')

## add x and y labels
## fontsize, changes the size of the labels
plt.xlabel("Sepal Width (cm)", fontsize=22)
plt.ylabel("Sepal Length (cm)", fontsize=22)

## adjusts the size of the tick labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

## adds a legend
plt.legend(fontsize=16)

## trim border whitespace
plt.tight_layout()

## saving the photo as a pdf
plt.savefig("iris_sepal_length_width.pdf",
               format='pdf')

### Alternatively ###

## saving the photo as an eps
plt.savefig("iris_sepal_length_width.eps",
               format='eps')


plt.show()
```

For more fine-grained control over your plots borders check out `subplots_adjust`, <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html">https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html</a>.
